```python
import numpy as np
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.proportion import proportions_ztest, proportion_effectsize
from statsmodels.stats.power import NormalIndPower

np.random.seed(42)

observed = np.array([120, 85, 110, 85]) 

total_players = 400
expected = np.array([100, 100, 100, 100]) 

chi2_stat, p_value = st.chisquare(observed, f_exp=expected)

print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"Degrees of Freedom: {len(observed) - 1}")
print(f"p-value: {p_value:.4f}")

if p_value < 0.05:
    print("Decision: Reject H₀. The character class distribution is not equal.")
else:
    print("Decision: Fail to reject H₀. There is no evidence that the distribution is not equal.")
```

### TASK 2

```python
data = np.array([[90, 60, 30], 
                 [110, 140, 170]])
df_table = pd.DataFrame(data, index=['Churned', 'Retained'], 
                        columns=['Free', 'Basic', 'Premium'])


chi2, p, dof, expected_table = st.chi2_contingency(data)

expected_df = pd.DataFrame(expected_table, index=df_table.index, columns=df_table.columns)
print("Expected Frequencies:")
print(expected_df)


print(f"\nChi-square: {chi2:.4f}\np-value: {p:.6f}\ndf: {dof}")
```

**Decision :** At $\alpha = 0.05$, we reject the null hypothesis ($H_0$) because the p-value is less than 0.05.Interpretation: There is statistically significant evidence of an association between subscription tier and churn status. This indicates that a customer's likelihood of leaving the service depends on whether they are on a Free, Basic, or Premium plan.

### TASK 3

```python
def cramers_v(chi2, n, min_dim):
    """Cramér's V """
    return np.sqrt(chi2 / (n * min_dim))

n_total = 600
v_result = cramers_v(chi2, n_total, 1)

print(f"Cramér's V: {v_result:.4f}")
```

**Effect Size Interpretation:** The relationship between subscription tier and churn is moderate (Cramér's V $\approx 0.30$)

**Marketing Implications:** This means the marketing team should not use a "one-size-fits-all" strategy. Since tier is a significant predictor of churn, they should design tier-specific retention campaigns, focusing especially on the Free and Basic segments where the risk of leaving is higher.
